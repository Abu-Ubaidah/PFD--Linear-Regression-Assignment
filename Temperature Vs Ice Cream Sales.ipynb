### https://www.kaggle.com/datasets/raphaelmanayon/temperature-and-ice-cream-sales/data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from pandas.core.common import random_state
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('Ice Cream Sales - temperatures.csv')
df.head()
```

```python
df.info()
```

```python
plt.title('Ice Cream Sales With Increasing Temperature')
sns.distplot(df['Ice Cream Profits'])
plt.show()
```

```python
plt.scatter(df['Temperature'], df['Ice Cream Profits'], color = 'lightcoral')
plt.title('Temperature vs Ice Cream Sales')
plt.xlabel('Temperature')
plt.ylabel('Ice Cream Profit')
plt.box(False)
plt.show()
```

```python
X = df[['Temperature']]  
y = df['Ice Cream Profits'] 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred_test = regressor.predict(X_test)
y_pred_train = regressor.predict(X_train) 
```

```python
plt.scatter(X_train, y_train, color = 'lightcoral')
plt.plot(X_train, y_pred_train, color = 'firebrick')
plt.title('Temperature vs Ice Cream Sales (Training Set)')
plt.xlabel('Temperature')
plt.ylabel('Ice Cream Sales')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Temperature/Ice Cream Sales', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
plt.scatter(X_test, y_test, color = 'lightcoral')
plt.plot(X_train, y_pred_train, color = 'firebrick')
plt.title('Temperatue Vs Ice Cream Sales (Test Set)')
plt.xlabel('Temperature')
plt.ylabel('Ice Cream Sales')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Temperature/Ice Cream Sales', loc='best', facecolor='white')
plt.box(False)
plt.show()
```
